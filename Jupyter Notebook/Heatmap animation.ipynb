```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation

# Load Data
data = pd.read_csv("..\Question 3_ Building Efficiency and Health\Sensor Readings\Group 3 RT.csv")

# Load Room Layout Image
room_map = plt.imread('..\Question 3_ Building Efficiency and Health\layout.png')

# Animated Heatmap
fig, ax = plt.subplots()
sns.set()

def animate(i):
    ax.clear()
    ax.imshow(room_map, extent=[0, 10, 0, 10])  # Adjust extent according to your room layout
sns.heatmap(data[data['Timestamp'] == i*2].pivot('Timestamp', 'X', 'Temperature'), ax=ax, cmap='coolwarm')

ani = animation.FuncAnimation(fig, animate, frames=len(data['Timestamp'].unique())//2, interval=200)

# To save the animation as a video file
ani.save('animated_heatmap.mp4', writer='ffmpeg')

# To display animation in notebook
plt.show()
```
